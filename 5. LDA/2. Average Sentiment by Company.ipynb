```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import csv

input_file = "lda.csv"
output_file = "company.csv"

with open(input_file, "r", encoding="utf-8", errors="replace") as file_in, open(output_file, "w", encoding="utf-8", newline="") as file_out:
    reader = csv.reader(file_in)
    writer = csv.writer(file_out)
    
    for row in reader:
        try:
            writer.writerow(row)
        except UnicodeDecodeError:
            continue
```

```python
data = pd.read_csv("company.csv")
```

```python
company_ratings = data.groupby('company')['rating'].mean()
```

```python
color = '#FFBE7A'
plt.bar(company_ratings.index, company_ratings.values,color=color)
#plt.grid(color='white', linestyle='--')
plt.rcParams['axes.facecolor'] = '#F5F5F530'
plt.xlabel('Company')
plt.ylabel('Average Rating')
plt.title('Average Sentiment by Company')
for i, value in enumerate(company_ratings.values):
    plt.text(i, value, str(round(value, 2)), ha='center', va='bottom')

plt.show()
```
