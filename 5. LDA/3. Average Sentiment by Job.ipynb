```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("techmean.csv")
mean_rating = data.groupby("Tech")["rating"].mean()

colors = ["#FFBE7A", "#82B0D2"]
bar_width = 0.5


labels = ["Non-Tech", "Tech"]
plt.bar(labels, mean_rating, width=bar_width, color=colors)

plt.xlabel("Tech")
plt.ylabel("Average Rating")
plt.title("Average Rating for Non-Tech and Tech")
plt.rcParams['axes.facecolor'] = '#F5F5F530'

for i, value in enumerate(mean_rating):
    plt.text(i, value, str(round(value, 2)), ha='center')

plt.show()
```
