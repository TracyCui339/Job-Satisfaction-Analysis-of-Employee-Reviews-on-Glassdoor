```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('timedata.csv')
df_mean = df.groupby('Year')['label'].mean().reset_index()

print(df_mean)


data = pd.read_csv("timedata.csv") 

df_mean = data.groupby('Year')['label'].mean().reset_index()

color = '#82B0D2'
color2 = '#999999'
color3='#FFBE7A'
plt.rcParams['axes.facecolor'] = '#99999920'
plt.plot(df_mean['Year'], df_mean['label'], color=color3, linewidth=2.5)
plt.grid(color='white', linestyle='--')
plt.title('Predicted Average Sentiment Change Trend', fontweight='bold', fontsize=14)
plt.xlabel('Year', fontweight='bold', fontsize=12)
plt.ylabel('Predicted Average Sentiment Average', fontweight='bold', fontsize=12)

plt.show()

for index, row in df_mean.iterrows():
    plt.text(row['Year'], row['label'], str(round(row['label'], 2)), ha='center', va='bottom')

plt.show()
```
