```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 读取csv文件
df = pd.read_csv('timedata.csv')

# 将Year_Month转换为日期格式
df['Yearmonth'] = pd.to_datetime(df['Yearmonth'], format='%Y-%m')
```

```python
df['Yearmonth']
```

```python
# 按月份计算情绪预测值的平均值
df_monthly = df.groupby('Yearmonth')['label'].mean().reset_index()

# 将Year_Month作为索引
df_monthly.set_index('Yearmonth', inplace=True)
```

```python
df_monthly
```

```python
# 绘制时间序列图
color = '#FFBE7A'
plt.plot(df_monthly,color=color)
plt.grid(color='white', linestyle='--')
plt.xlabel('Year_Month')
plt.ylabel('Predicted Average Sentiment')
plt.title('Predicted Average Sentiment Time Series')
plt.rcParams['axes.facecolor'] = '#F5F5F5'
plt.show()
```

```python
# 差分后绘制时间序列图
df_monthly_diff = df_monthly.diff().dropna()
plt.plot(df_monthly_diff, color=color)
plt.grid(color='white', linestyle='--')
plt.xlabel('Year_Month')
plt.ylabel('Predicted Average Sentiment Difference')
plt.title('Predicted Average Sentiment Time Series After Difference')
plt.rcParams['axes.facecolor'] = '#F5F5F5'
plt.show()
```

```python
# 绘制ACF图和PACF图
fig, ax = plt.subplots(2, 1, figsize=(10, 8))
sm.graphics.tsa.plot_acf(df_monthly_diff, lags=20, ax=ax[0])
sm.graphics.tsa.plot_pacf(df_monthly_diff, lags=20, ax=ax[1])
plt.show()
```

```python
# 拟合ARIMA模型并进行预测
from pmdarima.arima import auto_arima

model = auto_arima(df_monthly, seasonal=False, trace=True)
print(model.summary())
```

```python
import pandas as pd
import arch

# 创建ARCH/GARCH模型对象
model = arch.arch_model(df_monthly, vol='Garch', p=1, q=1)

# 拟合模型
model_fit = model.fit()

# 打印模型摘要信息
print(model_fit.summary())

# 进行40期的预测
forecast = model_fit.forecast(start=model_fit.conditional_volatility[-1:].index[0], horizon=40)

# 提取预测值和置信区间
forecast_values = forecast.mean.values[-40:]
conf_int = forecast.variance.values[-40:]

# 打印预测值和置信区间
for i in range(len(forecast_values)):
    print(f"期数{i+1}: 预测值={forecast_values[i]}, 置信区间={conf_int[i]}")
```

```python
history_length = len(df_monthly)

# 获取最新的数据用于预测
recent_data = df_monthly[-history_length:]

# 进行40期的预测
forecast_values, conf_int = model.predict(n_periods=3, return_conf_int=True, exogenous=recent_data)

# 打印预测值和置信区间
for i in range(len(forecast_values)):
    print(f"期数{i+1}: 预测值={forecast_values[i]}, 置信区间={conf_int[i]}")
```
