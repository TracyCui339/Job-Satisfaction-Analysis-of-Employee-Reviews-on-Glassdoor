```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 读取csv文件
df = pd.read_csv('new_predictdata.csv')

# 将Year_Month转换为日期格式
df['Year_Month'] = pd.to_datetime(df['Year_Month'], format='%Y-%m')
```

```python
df['Year_Month']
```

```python
# 按月份计算情绪预测值的平均值
df_monthly = df.groupby('Year_Month')['Sentiment'].mean().reset_index()

# 将Year_Month作为索引
df_monthly.set_index('Year_Month', inplace=True)
```

```python
df_monthly
```

```python
# 绘制时间序列图
color = '#FFBE7A'
plt.plot(df_monthly,color=color)
plt.grid(color='white', linestyle='--')
plt.xlabel('Year_Month')
plt.ylabel('Predicted Polarity Average')
plt.title('Predicted Polarity Average Time Series')
plt.rcParams['axes.facecolor'] = '#F5F5F5'
plt.show()
```

```python
# 差分后绘制时间序列图
df_monthly_diff = df_monthly.diff().dropna()
plt.plot(df_monthly_diff, color=color)
plt.grid(color='white', linestyle='--')
plt.xlabel('Year_Month')
plt.ylabel('Predicted Polarity Average Difference')
plt.title('Predicted Polarity Average Time Series After Difference')
plt.rcParams['axes.facecolor'] = '#F5F5F5'
plt.show()
```

```python
# 绘制ACF图和PACF图
fig, ax = plt.subplots(2, 1, figsize=(10, 8))
sm.graphics.tsa.plot_acf(df_monthly_diff, lags=20, ax=ax[0])
sm.graphics.tsa.plot_pacf(df_monthly_diff, lags=20, ax=ax[1])
plt.show()
```

```python
# 拟合ARIMA模型并进行预测
from pmdarima.arima import auto_arima

model = auto_arima(df_monthly_diff, seasonal=False, trace=True)
print(model.summary())
```
