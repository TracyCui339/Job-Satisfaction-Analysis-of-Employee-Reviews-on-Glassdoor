```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv("all_data.csv")

# 定义分类函数
def classify_aspect(row):
    income_keywords = ["middle income", "high income", "low income", "income inequality",
                       "disposable income", "household annual income", "wealth tax", "unemployment insurance"]
    occupation_keywords = ["meritocracy", "occupational mobility", "social mobility", "occupational prestige",
                           "parental occupation", "professional conversion programme", "skills future"]
    education_keywords = ["elite education", "education subsidies", "mendaki", "cdac", "less educated",
                          "educational attainment", "achievement gap", "access to education"]
    health_keywords = ["health insurance", "medical expenses", "medical resources", "medisave",
                       "medishield life", "community health assist scheme", "access to healthcare"]
    general_keywords = ["upper class", "lower class", "middle class", "central provident fund",
                        "cpf", "low income group", "middle income group", "high income group", "poverty"]

    # 将搜索术语转换为小写字母
    search_terms = row["Search_Term"].lower()
    
    # 检查每个关键词是否在“Search_Term”列中出现
    if any(keyword in search_terms for keyword in income_keywords):
        return "Income"
    elif any(keyword in search_terms for keyword in occupation_keywords):
        return "Occupation"
    elif any(keyword in search_terms for keyword in education_keywords):
        return "Education"
    elif any(keyword in search_terms for keyword in health_keywords):
        return "Health"
    elif any(keyword in search_terms for keyword in general_keywords):
        return "General"
    else:
        return "Other"


# 应用分类函数并添加“aspect”列
df["aspect"] = df.apply(classify_aspect, axis=1)

# 保存文件
df.to_csv("all_data2.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# 读取CSV文件
df = pd.read_csv("all_data2.csv")

# 将时间戳转换为年份
df["Year"] = pd.to_datetime(df["Timestamp"]).dt.year

# 按照aspect和年份计算sentiment的平均值
df_sentiment = df.groupby(["aspect", "Year"]).agg({"Sentiment": "mean"}).reset_index()

# 自定义颜色盘
colors = ['#8ECFC9', '#FFBE7A','#BB9727','#82B0D2','#BEB8DC','#E7DAD2','#2878B5']
cmap = ListedColormap(colors)

# 绘制每种Aspect的Sentiment平均值年份变化趋势
for i, aspect in enumerate(df_sentiment["aspect"].unique()):
    df_aspect = df_sentiment[df_sentiment["aspect"] == aspect]
    plt.plot(df_aspect["Year"], df_aspect["Sentiment"], label=aspect, color=cmap(i),linewidth=2.5)

# 添加图例和标题
plt.legend()
plt.rcParams['axes.facecolor'] = '#F5F5F5'
plt.grid(color='white', linestyle='--')
plt.title("Aspect Sentiment Trend")

# 显示图形
plt.show()
```
